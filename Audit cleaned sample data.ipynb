```python
#Connect to cleaned_sample.db using sqlite3.
import sqlite3

db = sqlite3.connect("C:\\Users\\JehyeonHeo\\sqlite_windows\\small_samples.db")
c = db.cursor()
```

```python
#query about postcodes in cleaned files.
query = "select tags.value, count(*) as count \
        from (select * from nodes_tags \
                union all \
                select * from ways_tags) tags \
        where tags.key = 'postcode' \
        group by tags.value \
        order by count desc;"
c.execute(query)
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#query about postcodes in uncleaned files.
query = "select tags.value, count(*) as count \
        from (select * from uncleaned_nodes_tags \
                union all \
                select * from uncleaned_ways_tags) tags \
        where tags.key = 'postcode' \
        group by tags.value \
        order by count desc;"
c.execute(query)
rows = c.fetchall()
for row in rows:
    print(row)
```
